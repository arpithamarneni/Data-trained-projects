```python
plt.figure(figsize=(20,5))
chart=sns.barplot(data=data,x="Area_Name",y="Cases_Property_Stolen")
chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
chart
```

Most of the property stolen cases are registed in Madhya Pradesh
