```python
# checking the state wise confirmed cases
plt.figure(figsize=(20,8))
sns.barplot(x="Province_State",y="People_Tested",data=covid)
plt.title("state wise People tested")
plt.xlabel("Province_State")
plt.xticks(rotation=45)
plt.ylabel("People_Tested")
```
